```python
import numpy as np
from scipy import linalg
```

```python
# a. Define matrix A
A = np.array([[1, -2, 3],
              [4,  5, 6],
              [7,  1, 9]])

# b. Define vector b
b = np.array([1, 2, 3])

# c. Solve the linear system A x = b
x = linalg.solve(A, b)
print(f"Solution x: {x}")
```

```python
# d. Check that the solution is correct
is_correct = np.allclose(A.dot(x), b)
print(f"Is the solution correct? {is_correct}")
```

```python
# e. Repeat using a random 3x3 matrix B
B = np.random.rand(3, 3)
X = linalg.solve(A, B)
print(f"Solution matrix X for A X = B:\n{X}")
```

```python
# f. Solve the eigenvalue problem for A
eigenvalues, eigenvectors = linalg.eig(A)
print(f"Eigenvalues:\n{eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")
```

```python
# g. Calculate the inverse and determinant of A
A_inv = linalg.inv(A)
A_det = linalg.det(A)
print(f"Inverse of A:\n{A_inv}")
print(f"Determinant of A: {A_det:.4f}")
```

```python
# h. Calculate the norm of A with different orders
norm_frob = linalg.norm(A)          # Frobenius norm
norm_1 = linalg.norm(A, 1)          # L1 norm
norm_inf = linalg.norm(A, np.inf)   # L-infinity norm
print(f"Frobenius Norm: {norm_frob:.4f}, L1 Norm: {norm_1}, L-inf Norm: {norm_inf}")
```

## a. Coordinate Normalization (Broadcasting)

```python
import numpy as np

p = np.array([[1.0, 2.0, 10],
              [3.0, 4.0, 20],
              [5.0, 6.0, 30],
              [7.0, 8.0, 40]])

z = p[:, 2]

z_col = z[:, np.newaxis] 
p_norm = p / z_col

print(p_norm)
```

## b. Fancy Indexing the Diagonal

```python
A = np.arange(9).reshape(3, 3)

diagonal = A[[0, 1, 2], [0, 1, 2]]

print(diagonal)
```

## c. Finding the Closest Number

```python
X = np.random.rand(10, 3)

distances = np.abs(X - 0.75)

j = np.argmin(distances, axis=1)

closest_values = X[np.arange(10), j]

print("Indices:", j)
print("Values:", closest_values)
```

## d. Predicting Slicing Shapes

```python
x = np.empty((10, 8, 6))

idx0 = np.zeros((3, 8)).astype(int)
idx1 = np.zeros((3, 1)).astype(int)
idx2 = np.zeros((1, 1)).astype(int)

print(x[idx0, idx1, idx2].shape) 
```

## e. Very Advanced 'as_strided'

```python
from numpy.lib.stride_tricks import as_strided

x = np.arange(12, dtype=np.int32).reshape((3, 4))

row_stride, col_stride = x.strides

z = as_strided(x, 
               shape=(2, 3, 2, 2),
               strides=(row_stride, col_stride, row_stride, col_stride))

print(z)
```
